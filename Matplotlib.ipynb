```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
language = ('Englis','French','Arabic','japanese')
pepole = (100,30,80,20)
ax.bar(language,pepole)
plt.xlabel("Language")
plt.ylabel("Pepole")
plt.show()
```

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
Languages = ("English","Arabic","Bangla","German","Japanese")
People = (100,70,20,30,25)
plt.pie(People, labels = Languages,autopct="%.1f%%")
plt.show()
```

```python
x = np.linspace(0,10,30)
y = np.sin(x)
z = np.cos(x)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(x,y , color = 'g')
ax.scatter(x,z ,color = 'b')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig3 = plt.figure()
ax = plt.add_subplot(projection='3d')

x = np.linspace(0, 2 * np.pi, 30)  # Define x values
z = 20 * np.random.random(30)
y = np.cos(x)
x = np.sin(x)

ax.scatter(x, y, z, c=z, cmap='viridis')  # Use a valid colormap
plt.show()
```
